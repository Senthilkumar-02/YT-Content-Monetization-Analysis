```python
X=df.drop('ad_revenue_usd',axis=1) # independent variable / Features / X
Y=df['ad_revenue_usd'] # dependent variable / Target/Label / y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, OneHotEncoder

# Define columns based on scaling strategy
minmax_cols = ["watch_time_minutes"]        # wide range → MinMaxScaler
robust_cols = ["subscribers"]               # heavy outliers → RobustScaler
standard_cols = ["likes", "comments"]       # normal-like → StandardScaler
# engagement_rate already small (0–0.25), no scaling needed

# Initialize scalers
scaler_minmax = MinMaxScaler()
scaler_robust = RobustScaler()
scaler_standard = StandardScaler()

# Work on copies (avoid modifying original data)
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Apply RobustScaler
X_train_scaled[robust_cols] = scaler_robust.fit_transform(X_train[robust_cols])
X_test_scaled[robust_cols] = scaler_robust.transform(X_test[robust_cols])

# Apply MinMaxScaler
X_train_scaled[minmax_cols] = scaler_minmax.fit_transform(X_train[minmax_cols])
X_test_scaled[minmax_cols] = scaler_minmax.transform(X_test[minmax_cols])

# Apply StandardScaler
X_train_scaled[standard_cols] = scaler_standard.fit_transform(X_train[standard_cols])
X_test_scaled[standard_cols] = scaler_standard.transform(X_test[standard_cols])

# Identify remaining columns
numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns.difference(
    robust_cols + minmax_cols + standard_cols
)
categorical_cols = X.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num_standard', StandardScaler(), numeric_cols),
        ('num_robust', RobustScaler(), robust_cols),
        ('num_minmax', MinMaxScaler(), minmax_cols),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols)
    ]
)
```

Linear Regression

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

pipe_lr = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

pipe_lr.fit(X_train, y_train)

y_train_pred_lr = pipe_lr.predict(X_train)
y_test_pred_lr = pipe_lr.predict(X_test)

print('LinearRegression Evaluation')
print('Train r2_score:',r2_score(y_train,y_train_pred_lr))
print('Test r2_score:',r2_score(y_test,y_test_pred_lr))
print('***'*10)
print('Train mean_squared_error:',mean_squared_error(y_train,y_train_pred_lr))
print('Test mean_squared_error:',mean_squared_error(y_test,y_test_pred_lr))
print('***'*10)
print('Train mean_absolute_error:',mean_absolute_error(y_train,y_train_pred_lr))
print('Test mean_absolute_error:',mean_absolute_error(y_test,y_test_pred_lr))
print('***'*10)
#7s
```

Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression

pipe_lr_poly = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('select', SelectKBest(score_func=f_regression, k=100)),  # keep best 100 features
    ('model', LinearRegression())
])

pipe_lr_poly.fit(X_train, y_train)

y_train_pred_poly = pipe_lr_poly.predict(X_train)
y_test_pred_poly = pipe_lr_poly.predict(X_test)

print('PolynomialFeatures Evaluation')
print('Train r2_score:',r2_score(y_train,y_train_pred_poly))
print('Test r2_score:',r2_score(y_test,y_test_pred_poly))
print('***'*10)
print('Train mean_squared_error:',mean_squared_error(y_train,y_train_pred_poly))
print('Test mean_squared_error:',mean_squared_error(y_test,y_test_pred_poly))
print('***'*10)
print('Train mean_absolute_error:',mean_absolute_error(y_train,y_train_pred_poly))
print('Test mean_absolute_error:',mean_absolute_error(y_test,y_test_pred_poly))
print('***'*10)
#5s
```

```python
# Collect metrics in a dictionary
results = {
    "Model": [],
    "Train R2": [],
    "Test R2": [],
    "Train MSE": [],
    "Test MSE": [],
    "Train MAE": [],
    "Test MAE": []
}
```

```python
# Helper function to append results
def add_results(model_name, y_train, y_train_pred, y_test, y_test_pred):
    results["Model"].append(model_name)
    results["Train R2"].append(r2_score(y_train, y_train_pred))
    results["Test R2"].append(r2_score(y_test, y_test_pred))
    results["Train MSE"].append(mean_squared_error(y_train, y_train_pred))
    results["Test MSE"].append(mean_squared_error(y_test, y_test_pred))
    results["Train MAE"].append(mean_absolute_error(y_train, y_train_pred))
    results["Test MAE"].append(mean_absolute_error(y_test, y_test_pred))

# Add each model’s results
add_results("LinearRegression", y_train, y_train_pred_lr, y_test, y_test_pred_lr)
add_results("Polynomial Regression", y_train, y_train_pred_poly, y_test, y_test_pred_poly)
add_results("Lasso", y_train, y_train_pred_lasso, y_test, y_test_pred_lasso)
add_results("Ridge", y_train, y_train_pred_ridge, y_test, y_test_pred_ridge)
add_results("ElasticNet", y_train, y_train_pred_elastic, y_test, y_test_pred_elastic)
add_results("RandomForest", y_train, y_train_pred_rf, y_test, y_test_pred_rf)
add_results("GradientBoosting", y_train, y_train_pred_gb, y_test, y_test_pred_gb)

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Sort by best Test R2 (higher is better)
results_df = results_df.sort_values(by="Test R2", ascending=False).reset_index(drop=True)

results_df
```
